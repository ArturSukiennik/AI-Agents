```python
from dotenv import load_dotenv
load_dotenv(override=True)

import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f'OpenAI API Key found {openai_api_key[:8]}...')
else:
    print('No OpenAI API Key found')
    
from openai import OpenAI
openai = OpenAI()

q1 = "Please pick a business area that might bew worth exploring for an Agentic AI oportunity. Specify it in a single sentence"
messages = [{'role': 'user', 'content': q1}]
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
business_area = response.choices[0].message.content

q2 = f"Please present a pain-point in the industry- something challenging that might be ripe for an Agentic AI solution. The industry is as follows: {business_area}."

messages = [{'role': 'user', 'content': q2}]
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
pain_point = response.choices[0].message.content
    
q3 = f"Propose Agentic AI solution in few lines for the given issues: {pain_point}"
messages = [{'role': 'user', 'content': q3}]
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
solution = response.choices[0].message.content
    
    
from IPython.display import Markdown, display
#display(Markdown(q1))
display(Markdown(business_area))

#display(Markdown(q2))
display(Markdown(pain_point))    

#display(Markdown(q3))
display(Markdown(solution))    
```
